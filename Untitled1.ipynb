```python
import pandas as pd
import os
from PIL import Image

# Đường dẫn tới CSV và folder ảnh
csv_url = "/content/drive/MyDrive/Banner/train/train/info.csv"
image_folder = "/content/drive/MyDrive/Banner/train/train/images"

# Đọc file CSV
data = pd.read_csv(csv_url)

# Tạo danh sách đường dẫn ảnh và captions
image_paths = [os.path.join(image_folder, img_name) for img_name in data['bannerImage']]
captions_list = data['caption'].tolist()

# Kiểm tra dữ liệu
print(f"Số lượng ảnh: {len(image_paths)}")
print(f"Caption đầu tiên: {captions_list[0]}")
print(f"Đường dẫn ảnh đầu tiên: {image_paths[0]}")
```

```python
from torch.utils.data import Dataset
from torchvision import transforms

# Dataset
class CustomDataset(Dataset):
    def __init__(self, image_paths, captions, transform=None):
        self.image_paths = image_paths
        self.captions = captions
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image = Image.open(self.image_paths[idx]).convert("RGB")
        caption = self.captions[idx]
        if self.transform:
            image = self.transform(image)
        return {"image": image, "caption": caption}

# Transform cho ảnh
transform = transforms.Compose([
    transforms.Resize((512, 512)),
    transforms.ToTensor()
])

# Tạo dataset
dataset = CustomDataset(image_paths, captions_list, transform=transform)
```

```python
from diffusers import StableDiffusionPipeline, StableDiffusionDreamBoothPipeline
from accelerate import Accelerator

# Đường dẫn tới thư mục chứa ảnh và thư mục lưu mô hình
instance_images_dir = "./instance_images"
output_dir = "./dreambooth_model"

# Từ khóa mô tả ảnh (phải độc đáo)
instance_prompt = "giày_dabo"  # Mô tả độc đáo cho đôi giày

# Tải mô hình pretrained
pretrained_model_name_or_path = "CompVis/stable-diffusion-v1-4"

# Khởi tạo pipeline
pipeline = StableDiffusionDreamBoothPipeline.from_pretrained(pretrained_model_name_or_path)
pipeline.to("cuda")

# Cấu hình huấn luyện
pipeline.train(
    instance_data_dir="/content/drive/MyDrive/Banner/train/train/images",
    instance_prompt=instance_prompt,
    output_dir=output_dir,
    resolution=512,
    train_batch_size=1,
    num_train_epochs=5,
    gradient_accumulation_steps=1,
    learning_rate=5e-6,
    lr_scheduler="constant",
    lr_warmup_steps=0,
    save_steps=500,
)
```

```python
prompt = "Đôi giày da cao cấp dành cho quý ông, đẳng cấp và lịch lãm"
generated_image = pipeline(prompt).images[0]

# Lưu ảnh
generated_image.save("generated_image.png")
generated_image.show()
```
